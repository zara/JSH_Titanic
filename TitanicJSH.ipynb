```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import pandas as pd
import scipy as sp
```

```python
def jsd(p,q,base=np.e):
    p,q=np.asarray(p), np.asarray(q)
    p,q=p/p.sum(), q/q.sum()
    m=1./2*(p+q)
    return sp.stats.entropy(p,m,base=base)/2. + sp.stats.entropy(q,m,base=base)/2
```

```python
df = pd.read_csv("titanic.csv")
df['Age'].fillna(df['Age'].median(), inplace=True)
males_Age=df['Age'][df['Sex'].isin(["male"])]
females_Age=df['Age'][df['Sex'].isin(["female"])]
p, devision = np.histogram(males_Age, bins=300)
q, devision = np.histogram(females_Age, bins=300)
plt.title('JS')
print(jsd(p,q))
plt.plot(p, c='blue')
plt.plot(q, c='red')
```
